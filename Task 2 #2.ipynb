```python
import pandas as pd
from sklearn.cluster import KMeans 
from matplotlib import pyplot as plt

# Step 1: Read the Data

d=pd.read_csv('Iris.csv')
print(d.head(10))
```

```python

x=d.iloc[:,1:5].values
 
#Step 2: Build the Model
    
km=KMeans(n_clusters=10)
km.fit(x)
yk=km.predict(x)
plt.scatter(x[:, 0], x[:, 1],c=yk, s=30, cmap='inferno')
c=km.cluster_centers_
plt.scatter(c[:,0], c[:,1],c="black", s=200,alpha=0.7)
```

```python
# Step 3: Optimum clusters

sse=[]
for i in range(1,11):
    m=KMeans(n_clusters=i)
    m.fit(x)
    sse.append(m.inertia_)
plt.plot(range(1,11),sse)
```

```python
# Step 4: Plot

km3=KMeans(n_clusters=3)
km3.fit(x)
pcl=km3.fit_predict(x)
c3=km3.cluster_centers_
print(c3)
#plt.scatter(x[:,0],x[:,1],c=pcl,cmap='viridis')
plt.scatter(x[yk==0,0], x[yk==0,1],s=50,c="black")
plt.scatter(x[yk==1,0], x[yk==1,1],s=50,c="green")
plt.scatter(x[yk==2,0], x[yk==2,1],s=50,c="blue")
plt.scatter(c3[:,0],c3[:,1],s=100,c="red")
```
