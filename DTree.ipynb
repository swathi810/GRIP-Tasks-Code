```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 

# Step 1: read the data

data=pd.read_csv('Iris.csv')
data.head()
```

```python
datacleaned=data.drop("Id",axis=1)
datacleaned.head()
```

```python
datacleaned.describe
```

```python
datacleaned['SepalLengthCm'].plot.hist()
plt.grid()
plt.title("SepalLength Plot")
plt.show()
```

```python
# Step 2: Model Building

inputs=datacleaned[['SepalLengthCm','SepalLengthCm','PetalLengthCm','PetalWidthCm']].values
classes=datacleaned['Species'].values
tr_inp,test_inp,tr_class,test_class=train_test_split(inputs,classes,test_size=0.2)
dtree=DecisionTreeClassifier()
dtree.fit(tr_inp,tr_class)
print(dtree.score(test_inp,test_class))
pred=dtree.predict(test_inp)
```

```python
#Step 3: Visualization

from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dtree.fit(inputs,classes),filled=True,proportion=True,precision=4)
plt.show()
```
